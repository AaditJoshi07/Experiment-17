```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Line Chart (Trend)

plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.title('Study Hours Over Days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
#Advanced line chart

plt.figure(figsize=(7,4))

plt.plot(df['Days'],df['Study_Hours'],marker='o',label='Study Hours')
plt.plot(df['Days'],df['Marks'],marker='o',label='Marks')

plt.title('Study hours over days')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Bar Chart comparison

plt.bar(df['Days'],df['Marks'],color='red',edgecolor='black',label='Study Hours')
plt.title("Comparison")
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.grid(axis='y',linestyle='--',alpha=1)
plt.legend()
plt.show()
#
```

```python
#Advanced Bar Graph

bars = plt.bar(df['Days'],df['Marks'],color='red',edgecolor='black',label='Study Hours')
for bar in bars:
  y=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,y, str(y),ha='center',va='bottom')

plt.title("Comparison")
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y',linestyle='--',alpha=1)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.arange(len(df['Days']))
width = 0.35

plt.bar(x - width/2, df['Study_Hours'],width,label='Study Hours')
plt.bar(x + width/2, df['Marks'],width,label='Marks')

plt.xticks(x,df['Days'])
plt.title('Aadit Joshi')
plt.xlabel('Days')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# EXAMPLE data
marks=[50,60,55,70,65,75,80,85,90,45,67,72,77,82,88]
#Calculate Mean
mean_value = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=3,color='skyblue',edgecolor='black')
# Mean line
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label='Mean')
plt.title("Advanced Histogram of Marks")
plt.xlabel("Marks") # Removed space
plt.ylabel("Frequency") # Added y-label for clarity
plt.legend()
plt.grid(axis='y',linestyle='-.',alpha=0.5)
plt.show() # Added plt.show() to display the plot

```

```python
# Scatter Plot(relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title("Aadit")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
